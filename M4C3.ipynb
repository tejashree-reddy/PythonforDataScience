##### 1 plot a Bar Graph of the data, taking the Year as the x-axis and the number of hurricanes occurring as the Y-axis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv("Hurricanes.csv")

data.index = np.arange(1, len(data)+1)
data.head()
data.shape
data.tail()
```

##### 5.5: Give error bar if  y_error = [0.12, 0.13, 0.2, 0.1]
##### 5.6: define width, height as figsize=(4,5) DPI and adjust plot dpi=100
##### 5.7: Give a font size of 14


```python
%matplotlib inline
x = [1,2,3,4]
y = [20, 21, 20.5, 20.8]
plt.errorbar(x, y, color="c", yerr=[0.12, 0.13, 0.2,0.1])
# plt.show()
```

```python
plt.figure(figsize=(4,5),dpi=100)

fig, ax = plt.subplots()


ax.errorbar(x, y,
            yerr=[0.12, 0.13, 0.2,0.1],
            fmt='-o')


ax.set_xlabel('x-axis',fontsize=14)
ax.set_ylabel('y-axis',fontsize=14)
ax.set_title('Line plot with error bars',fontsize=14)


plt.show()
```

##### Draw a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age


```python
x=data["preTestScore"]
y=data["postTestScore"]
size=data["age"]
plt.scatter(x, y, s=size,c='red')
plt.title('Scatterplot of preTestScore and postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
```
