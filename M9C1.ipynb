```python
Udata=Gdata.drop(['kurt', 'centroid', 'dfrange'],axis=1)
```

```python
Udata.info()
```

```python
from sklearn import preprocessing 
```

```python
min_max_scaler = preprocessing.MinMaxScaler(feature_range =(0, 16))
```

```python
Y=Udata['label']
X=Udata.drop('label', axis=1)
```

```python
x_after_min_max_scaler = min_max_scaler.fit_transform(X) 
```

```python
pd.DataFrame(x_after_min_max_scaler)
```

```python
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
```

```python
model = SVC()
```

```python
trainx, testx, trainy, testy=train_test_split(x_after_min_max_scaler,Y,test_size=0.3)
```

```python
model.fit(trainx, trainy)
```

```python
predictedy = model.predict(testx)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(testy, predictedy)
```
