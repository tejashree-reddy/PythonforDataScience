```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##### 1.Compute how much total salary cost has increased from year 2011 to 2014

```python
sal=pd.read_csv("Salaries.csv")
sal
```

```python
sal.columns
sal.info()
```

```python
saldata=sal.loc[:,['Year','TotalPay']]
saldata
```

```python
increase_sal=saldata.groupby("Year").mean().reset_index()
increase_sal
```

```python
plt.bar(increase_sal["Year"],increase_sal["TotalPay"])
plt.xlabel("Year")
plt.ylabel("Salaries")
plt.show()
```

##### 2.Which Job Title in Year 2014 has highest mean salary?

```python
Job=sal.loc[sal["Year"]==2014]
Job
```

```python
J=Job.loc[:,["JobTitle","TotalPay","Year"]]
J
```

```python
jobSal=J.groupby("JobTitle").mean().reset_index()
jobSal
```

```python
jobSal[jobSal["TotalPay"]==jobSal["TotalPay"].max()]
```

##### 3.How much money could have been saved in Year 2014 by stopping OverTimePay?

```python
OT=Job.loc[:,["JobTitle","OvertimePay","TotalPay","Year"]]
OT
```

```python
OT["TotalPay"].sum()-OT["OvertimePay"].sum()
```

##### 2.In year 2011 OverTimePay was what percentage of  TotalPayBenefits

```python
data2011=sal.loc[sal["Year"]==2011]
data2011
```

```python
cal=data2011.loc[:,['OvertimePay','TotalPayBenefits']]
cal
```

```python
cal["OvertimePay"].sum()-cal["TotalPayBenefits"].sum()
```

```python
(cal["TotalPayBenefits"].sum()-cal["OvertimePay"].sum())/cal["TotalPayBenefits"].sum()
```

```python
((cal["TotalPayBenefits"].sum()-cal["OvertimePay"].sum())/cal["TotalPayBenefits"].sum())*100
```

```python
cal["percentage"]=(cal["TotalPayBenefits"]-cal["OvertimePay"])/cal["TotalPayBenefits"]
cal["percentage"]*100
```
