```python
import pandas as pd
import numpy as np
```

```python
Data_Train=pd.read_csv("train.csv")
Data_Test=pd.read_csv("test.csv")
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
Data_Train['species']= label_encoder.fit_transform(Data_Train['species'])
Data_Train['species'].nunique()
```

```python
Data_Train.head()
```

```python
X=Data_Train.iloc[:,2:195]
```

```python
Y=Data_Train.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
rfc=RandomForestClassifier()
svc=SVC()
dtc=DecisionTreeClassifier()
nb=GaussianNB()
```

```python
rfc=rfc.fit(x_train,y_train)
svc=svc.fit(x_train,y_train)
dtc=dtc.fit(x_train,y_train)
nb=nb.fit(x_train,y_train)
```

```python
y_pred_rfc=rfc.predict(x_test)
y_pred_svc=svc.predict(x_test)
y_pred_dtc=dtc.predict(x_test)
y_pred_nb=nb.predict(x_test)
```

```python
acc_rfc=metrics.accuracy_score(y_test, y_pred_rfc)
acc_svc=metrics.accuracy_score(y_test, y_pred_svc)
acc_dtc=metrics.accuracy_score(y_test, y_pred_dtc)
acc_nb=metrics.accuracy_score(y_test, y_pred_nb)
```
